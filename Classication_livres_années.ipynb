```python
categories =set(d['categories'] for d in books)
```

```python
print(categories)
```

```python
isbn =set(d['isbn'] for d in books)
```

```python
print(isbn)
```

```python
Isbn_categories = []
for t in categories: 
    sumbooks = 0
    categoriescount = 0
    for d in books:
        if d['categories'] == t:
            sumbooks += 1
            categoriescount += 1
    Isbn_categories.append((t, sumbooks /categoriescount))
```

```python
print(Isbn_categories)
```
